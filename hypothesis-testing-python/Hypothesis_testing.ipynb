Estimation And Confidence Intervals

Import Required Libraries

```python
import numpy as np
from scipy import stats
```

Enter the Sample Data

```python
# Durability data (in million characters)

data = [1.13, 1.55, 1.43, 0.92, 1.25,
        1.36, 1.32, 0.85, 1.07, 1.48,
        1.20, 1.33, 1.18, 1.22, 1.29]
```

Calculate Sample Statistics

```python
# Sample size
n = len(data)

# Sample mean
mean = np.mean(data)

# Sample standard deviation (ddof=1 gives sample SD)
sample_sd = np.std(data, ddof=1)

print("Sample Size =", n)
print("Sample Mean =", round(mean,4))
print("Sample Standard Deviation =", round(sample_sd,4))
```

Calculate the Confidence Interval

```python
# Confidence level
confidence = 0.99

# Degrees of freedom
df = n - 1

# t critical value
t_critical = stats.t.ppf((1 + confidence) / 2, df)

# Margin of Error
margin_error = t_critical * (sample_sd / np.sqrt(n))

# Confidence Interval
lower = mean - margin_error
upper = mean + margin_error

print("99% Confidence Interval using t-distribution")
print("Lower Limit =", round(lower,4))
print("Upper Limit =", round(upper,4))
```

Calculate Confidence Interval Using Z-distribution

```python

# Known population standard deviation
sigma = 0.2
z_critical = stats.norm.ppf(0.995)
margin_error_z = z_critical * (sigma / np.sqrt(n))

lower_z = mean - margin_error_z
upper_z = mean + margin_error_z

print("99% Confidence Interval using Z-distribution")
print("Lower Limit =", round(lower_z,4))
print("Upper Limit =", round(upper_z,4))
```

Hypothesis testing

Import Required Library

```python
import numpy as np
from scipy.stats import norm
```

Given Data

```python


sample_mean = 3050
x = 600
sample_size = 25

# Cost model: W = 1000 + 5X
theoretical_mean = 1000 + (5 * x)

# Standard deviation of X
sigma_x = 25

# Standard deviation of operating cost
sigma = 5 * sigma_x

alpha = 0.05
```

State the Hypotheses

```python
print("Hypothesis Testing")
print("------------------------------")
print("Null Hypothesis (H0): μ = 4000")
print("Alternative Hypothesis (H1): μ > 4000")
```

Calculate Test Statistic (Z-Test)

```python
import math

# Calculate Z statistic

z = (sample_mean - theoretical_mean) / (sigma / math.sqrt(sample_size))

print("\nTheoretical Mean =", theoretical_mean)
print("Standard Deviation =", sigma)
print("Calculated Z-value =", round(z,4))
```

Critical Value and P-value

```python
# Right-tailed test

critical_value = norm.ppf(1-alpha)

p_value = 1 - norm.cdf(z)

print("\nCritical Value =", round(critical_value,4))
print("P-value =", round(p_value,6))
```

Decision

```python
print("\nDecision")

if z > critical_value:
    print("Reject the Null Hypothesis (H0)")
else:
    print("Fail to Reject the Null Hypothesis (H0)")
```

Conclusion

```python
print("\nConclusion")

if z > critical_value:
    print("There is sufficient evidence to support the restaurant owners' claim that")
    print("weekly operating costs are higher than the theoretical cost model.")
else:
    print("There is NOT sufficient evidence to support the restaurant owners' claim.")
    print("The observed sample does not indicate higher operating costs than predicted by the model.")
```

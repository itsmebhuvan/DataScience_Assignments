6. Given a 'marks' variable, print the grade:- 90 and above: 'A'- 75–89: 'B'- 50–74: 'C'- Below 50: 'Fail'

```python
marks = int(input("Enter your marks"))
if marks >= 90:
    print("A")
elif marks >= 75:
    print("B")
elif marks >= 50:
    print("C")
else:
    print("Fail")
```

```python
num = int(input("Enter a number"))
if num > 0:
    if num % 2 == 0:
        print("Positive and even",num )
    else:
      print("Positive and odd",num)

elif num < 0:
  if num % 2 == 0:
    print("Negative and even",num)
  else:
     print("Negative and odd", num)

elif num == 0:
    print("Zero",num)
```

Handling Missing Values

```python
from sklearn.impute import SimpleImputer

df.replace(" ?", np.nan, inplace=True)
df.replace("?", np.nan, inplace=True)

num_cols = df.select_dtypes(include=np.number).columns
cat_cols = df.select_dtypes(exclude=np.number).columns

num_imputer = SimpleImputer(strategy="median")
cat_imputer = SimpleImputer(strategy="most_frequent")

df[num_cols] = num_imputer.fit_transform(df[num_cols])
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])
```

Feature Scaling

Standard Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df_standard = df.copy()
df_standard[num_cols] = scaler.fit_transform(df[num_cols])
```

Min-Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df_minmax = df.copy()
df_minmax[num_cols] = scaler.fit_transform(df[num_cols])
```

Load Dataset & Basic Data Exploration

```python
import pandas as pd

# Load dataset
df = pd.read_csv("/content/adult_with_headers (1).csv")

# First 5 rows
print(df.head())

# Dataset information
print(df.info())

# Summary statistics
print(df.describe())

# Missing values
print(df.isnull().sum())

# Check for '?' values (Adult dataset usually stores missing values as '?')
print((df == '?').sum())
```

Handle Missing Values

```python
import numpy as np

# Replace '?' with NaN
df.replace('?', np.nan, inplace=True)

# Check missing values
print(df.isnull().sum())

# Fill categorical columns with mode
for col in df.select_dtypes(include='object'):
    df[col].fillna(df[col].mode()[0], inplace=True)

# Fill numerical columns with median
for col in df.select_dtypes(include=['int64','float64']):
    df[col].fillna(df[col].median(), inplace=True)
```

. Scaling Numerical Features

Standard Scaling

```python
from sklearn.preprocessing import StandardScaler

standard_scaler = StandardScaler()

num_cols = [
    'age',
    'fnlwgt',
    'education_num',
    'capital_gain',
    'capital_loss',
    'hours_per_week'
]

standard_scaled = standard_scaler.fit_transform(df[num_cols])
```

Min-Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler()

minmax_scaled = minmax.fit_transform(df[num_cols])
```

One-Hot Encoding

```python
small_cat = [col for col in df.select_dtypes(include='object')
             if df[col].nunique() < 5]

onehot = pd.get_dummies(df, columns=small_cat)

print(onehot.head())
```

Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df_label = df.copy()

for col in df_label.select_dtypes(include='object'):
    df_label[col] = label_encoder.fit_transform(df_label[col])

print(df_label.head())
```

Task 2: Dimensionality Reduction using PCA

Standardize the Dataset

```python
from sklearn.preprocessing import StandardScaler

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

print("Shape of Scaled Data:", scaled_data.shape)
```

Apply PCA

```python
from sklearn.decomposition import PCA

# Apply PCA
pca = PCA()

pca_data = pca.fit_transform(scaled_data)
```

Explained Variance Ratio

```python
explained_variance = pca.explained_variance_ratio_

print(explained_variance)
```

Scree Plot

```python
plt.figure(figsize=(8,5))

plt.plot(
    range(1,len(explained_variance)+1),
    explained_variance,
    marker='o'
)

plt.xlabel("Principal Components")
plt.ylabel("Explained Variance")
plt.title("Scree Plot")

plt.grid(True)
plt.show()
```

Cumulative Explained Variance

```python
cumulative_variance = explained_variance.cumsum()

plt.figure(figsize=(8,5))

plt.plot(
    range(1,len(cumulative_variance)+1),
    cumulative_variance,
    marker='o'
)

plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")

plt.axhline(0.95,color='red',linestyle='--')

plt.grid(True)
plt.show()
```

Select Optimal Components

```python
optimal_components = np.argmax(cumulative_variance >= 0.95) + 1

print("Optimal Number of Components:", optimal_components)
```

Transform Dataset

```python
pca = PCA(n_components=optimal_components)

pca_dataset = pca.fit_transform(scaled_data)

print(pca_dataset.shape)
```

Task 3: Clustering with Original Data

Apply K-Means

```python
from sklearn.cluster import KMeans

kmeans_original = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

original_labels = kmeans_original.fit_predict(scaled_data)
```

Visualize Clusters

```python
plt.figure(figsize=(8,6))

plt.scatter(
    scaled_data[:,0],
    scaled_data[:,1],
    c=original_labels,
    cmap='viridis'
)

plt.xlabel(df.columns[0])
plt.ylabel(df.columns[1])

plt.title("Clusters on Original Dataset")

plt.show()
```

Evaluate Clustering

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score

silhouette_original = silhouette_score(
    scaled_data,
    original_labels
)

db_original = davies_bouldin_score(
    scaled_data,
    original_labels
)

print("Silhouette Score :", silhouette_original)

print("Davies-Bouldin Index :", db_original)
```

Task 4: Clustering with PCA Data

Apply K-Means

```python
kmeans_pca = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

pca_labels = kmeans_pca.fit_predict(pca_dataset)
```

Visualize PCA Clusters

```python
plt.figure(figsize=(8,6))

plt.scatter(
    pca_dataset[:,0],
    pca_dataset[:,1],
    c=pca_labels,
    cmap='rainbow'
)

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")

plt.title("Clusters after PCA")

plt.show()
```

Evaluate PCA Clustering

```python
silhouette_pca = silhouette_score(
    pca_dataset,
    pca_labels
)

db_pca = davies_bouldin_score(
    pca_dataset,
    pca_labels
)

print("Silhouette Score :", silhouette_pca)

print("Davies-Bouldin Index :", db_pca)
```

Task 5: Comparison and Analysis

Create Comparison Table

```python
comparison = pd.DataFrame({

    "Method":[
        "Original Dataset",
        "PCA Dataset"
    ],

    "Silhouette Score":[
        silhouette_original,
        silhouette_pca
    ],

    "Davies-Bouldin Index":[
        db_original,
        db_pca
    ]

})

comparison
```

Compare Using Bar Charts

```python
comparison.plot(
    x="Method",
    y="Silhouette Score",
    kind="bar",
    figsize=(6,5)
)

plt.title("Silhouette Score Comparison")

plt.show()
```

Davies-Bouldin Index

```python
comparison.plot(
    x="Method",
    y="Davies-Bouldin Index",
    kind="bar",
    figsize=(6,5)
)

plt.title("Davies-Bouldin Index Comparison")

plt.show()
```
